```python
import pandas as pd
import numpy as np

df = pd.read_csv('dota2Train.csv',header=None)
df_test = pd.read_csv('dota2Test.csv', header=None)

df.head()
```

```python
from sklearn.model_selection import cross_val_score

y = df[0]
X = df.iloc[:,1:]

y_test = df_test[0]
X_test = df_test.iloc[:,1:]

X.shape, y.shape
```

```python
# Perceptron
from sklearn.linear_model import perceptron

# Create the model
net = perceptron.Perceptron(max_iter=100)
net.fit(X,y)

# Results
predictions = net.predict(X_test)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y_test) )
print("Accuracy   ", str(net.score(X_test, y_test)*100) + "%" ) # mean accuracy on the given test data and labels.

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(perceptron.Perceptron(max_iter=100), 
                         X, y, scoring='accuracy', cv=5)
print('\n5-fold Cross validation results',scores)
print('Average from 10-fold cross validation:',scores.mean())
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

# Create the model
lin = LinearRegression()
lin.fit(X,y)

# Results
predictions = lin.predict(X_test)
 
# Print the results
print("Predictions", np.sign(predictions) )
print("Actual     ", np.matrix(y_test) )
print("R^2        ", str(lin.score(X_test, y_test)) ) # mean accuracy on the given test data and labels.

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(LinearRegression(), 
                         X, y, scoring='r2', cv=5)
print('\n5-fold Cross validation results',scores)
print('Average from 10-fold cross validation:',scores.mean())
```

```python
# Multi-Layer Perceptron

from sklearn.neural_network import MLPClassifier

# Create the model
net = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(10,), random_state=1)
net.fit(X,y)

# Results
predictions = net.predict(X_test)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y) )
print("Accuracy   ", str(net.score(X_test, y_test)*100) + "%" )

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5,2), random_state=1),
                         X, y, scoring='accuracy', cv=5)
print('\n5-fold Cross validation results',scores)
print('Average from 10-fold cross validation:',scores.mean())
```

```python
# SVM
from sklearn import svm

# Create the model
clf = svm.SVC(kernel='linear')
clf.fit(X,y)

# Results
predictions = clf.predict(X)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y) )
print("Accuracy   ", str(clf.score(X, y)*100) + "%" )

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(svm.SVC(kernel='linear'),
                         X, y, scoring='accuracy', cv=5)
print('\n10-fold Cross validation results',scores)
print('Average from 5-fold cross validation:',scores.mean())
```
