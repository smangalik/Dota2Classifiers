```python
import pandas as pd
import numpy as np

df = pd.read_csv('dota2Train.csv',header=None)

df.head()
```

```python
from sklearn.model_selection import cross_val_score

target = df[0]
features = df.iloc[:,1:]

target.shape, features.shape
```

```python
# Perceptron
from sklearn.linear_model import perceptron

# Train and Fit Model
y = target
X = features

# Create the model
net = perceptron.Perceptron(max_iter=None, tol=1e-4, shuffle=True)
net.fit(X,y)

# Results
predictions = net.predict(X)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y) )
print("Accuracy   ", str(net.score(X, y)*100) + "%" ) # mean accuracy on the given test data and labels.

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(perceptron.Perceptron(max_iter=None, tol=1e-4, shuffle=True), 
                         X, y, scoring='accuracy', cv=5)
print('\n5-fold Cross validation results',scores)
print('Average from 10-fold cross validation:',scores.mean())
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

# Train and Fit Model
y = target
X = features

# Create the model
lin = LinearRegression()
lin.fit(X,y)

# Results
predictions = lin.predict(X)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y) )
print("R^2        ", str(lin.score(X, y)) ) # mean accuracy on the given test data and labels.

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(LinearRegression(), 
                         X, y, scoring='r2', cv=5)
print('\n5-fold Cross validation results',scores)
print('Average from 10-fold cross validation:',scores.mean())
```

```python
# Multi-Layer Perceptron

from sklearn.neural_network import MLPClassifier

# Train and Fit Model
y = target
X = features

# Create the model
net = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(10,), random_state=1)
net.fit(X,y)

# Results
predictions = net.predict(X)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y) )
print("Accuracy   ", str(net.score(X, y)*100) + "%" )

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5,2), random_state=1),
                         X, y, scoring='accuracy', cv=5)
print('\n5-fold Cross validation results',scores)
print('Average from 10-fold cross validation:',scores.mean())
```

```python
# SVM
from sklearn import svm

# Train and Fit Model
y = target
X = features

# Create the model
clf = svm.SVC()
clf.fit(X,y)

# Results
predictions = clf.predict(X)
 
# Print the results
print("Predictions", predictions )
print("Actual     ", np.matrix(y) )
print("Accuracy   ", str(clf.score(X, y)*100) + "%" )

# evaluate the model using 10-fold cross-validation
scores = cross_val_score(svm.SVC(),
                         X, y, scoring='accuracy', cv=5)
print('\n10-fold Cross validation results',scores)
print('Average from 5-fold cross validation:',scores.mean())
```
